## Продвинутые методы

| Метод | Описание | Когда использовать |
|-------|----------|-------------------|
| **IA³** | Infused Adapter by Inhibiting/Amplifying | Меньше параметров, задача-specific |
| **DoRA** | Weight-Decomposed LoRA | Лучшее качество при том же rank |
| **Multitask** | Несколько адаптеров | Один базовый + много специализаций |
