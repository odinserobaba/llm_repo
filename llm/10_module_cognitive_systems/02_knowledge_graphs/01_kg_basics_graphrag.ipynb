## 1. –°–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∞ (networkx)

```python
import networkx as nx

# ‚Üê 1. –°–æ–∑–¥–∞—ë–º –≥—Ä–∞—Ñ
G = nx.DiGraph()

# ‚Üê 2. –î–æ–±–∞–≤–ª—è–µ–º —É–∑–ª—ã (—Å—É—â–Ω–æ—Å—Ç–∏)
G.add_node("–ò–≤–∞–Ω", type="PERSON")
G.add_node("–Ø–Ω–¥–µ–∫—Å", type="ORG")
G.add_node("–ú–æ—Å–∫–≤–∞", type="LOC")

# ‚Üê 3. –î–æ–±–∞–≤–ª—è–µ–º —Ä—ë–±—Ä–∞ (—Å–≤—è–∑–∏)
G.add_edge("–ò–≤–∞–Ω", "–Ø–Ω–¥–µ–∫—Å", relation="works_at")
G.add_edge("–Ø–Ω–¥–µ–∫—Å", "–ú–æ—Å–∫–≤–∞", relation="located_in")

# ‚Üê 4. –ü–æ–∏—Å–∫ –ø—É—Ç–µ–π
paths = list(nx.all_simple_paths(G, "–ò–≤–∞–Ω", "–ú–æ—Å–∫–≤–∞"))
print("–ü—É—Ç–∏ –ò–≤–∞–Ω ‚Üí –ú–æ—Å–∫–≤–∞:", paths)
#        ‚Üë
#        ‚îî‚îÄ‚îÄ ‚ú® –í–û–¢ –ó–î–ï–°–¨ –ú–ê–ì–ò–Ø: –ø–æ–∏—Å–∫ –ø–æ —Å–≤—è–∑—è–º, –∞ –Ω–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
```
