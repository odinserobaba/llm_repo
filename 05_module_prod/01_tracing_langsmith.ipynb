## –ù–∞—Å—Ç—Ä–æ–π–∫–∞ LangSmith

–ù—É–∂–Ω–æ —É–∫–∞–∑–∞—Ç—å –∫–ª—é—á LangSmith –∏ –≤–∫–ª—é—á–∏—Ç—å —Ç—Ä–∞—Å—Å–∏—Ä–æ–≤–∫—É.

> –í Colab –º–æ–∂–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å `LANGCHAIN_API_KEY` –≤ Secrets (üîë), —á—Ç–æ–±—ã –Ω–µ –≤–≤–æ–¥–∏—Ç—å –µ–≥–æ –∫–∞–∂–¥—ã–π —Ä–∞–∑.

```python
import os
from getpass import getpass
from dotenv import load_dotenv

load_dotenv()

# –í–∫–ª—é—á–∞–µ–º —Ç—Ä–∞—Å—Å–∏—Ä–æ–≤–∫—É
os.environ["LANGCHAIN_TRACING_V2"] = "true"

if not os.environ.get("LANGCHAIN_API_KEY"):
    os.environ["LANGCHAIN_API_KEY"] = getpass("–í–≤–µ–¥–∏—Ç–µ LANGCHAIN_API_KEY: ")
```

## –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø—Ä–æ–≤–∞–π–¥–µ—Ä–∞ (AITunnel)

–£–∫–∞–∂–µ–º –∫–ª—é—á –∏ –±–∞–∑–æ–≤—ã–π URL –¥–ª—è –≤—ã–∑–æ–≤–æ–≤ –º–æ–¥–µ–ª–∏.

```python
if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("–í–≤–µ–¥–∏—Ç–µ OPENAI_API_KEY: ")

if not os.environ.get("OPENAI_BASE_URL"):
    os.environ["OPENAI_BASE_URL"] = "https://api.aitunnel.ru/v1/"
```

## –ü—Ä–∏–º–µ—Ä —Ü–µ–ø–æ—á–∫–∏ —Å —Ç—Ä–∞—Å—Å–∏—Ä–æ–≤–∫–æ–π

–õ—é–±–æ–π –≤—ã–∑–æ–≤ —á–µ—Ä–µ–∑ LangChain —Ç–µ–ø–µ—Ä—å –±—É–¥–µ—Ç –ª–æ–≥–∏—Ä–æ–≤–∞—Ç—å—Å—è –≤ LangSmith.

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(
    model="gpt-5.2-chat",
    temperature=0.2,
    max_tokens=256,
    base_url=os.environ.get("OPENAI_BASE_URL"),
)

prompt = PromptTemplate.from_template(
    "–û–±—ä—è—Å–Ω–∏ –ø—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏, —á—Ç–æ —Ç–∞–∫–æ–µ RAG."
)

chain = prompt | llm | StrOutputParser()

result = chain.invoke({})
print(result)
```
